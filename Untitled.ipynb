```python
# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event distance/length         object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed         object
# Athlete ID                     int64
# Country tag                   object
# Athlete age                  float64
# dtype: object
```

```python
new_df = new_df.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_lenght',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlere_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id',
    'Athlete age': 'athlete_age'
                       
})
```

```python
final_df = new_df[['race_day', 'race_name', 'race_lenght', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlere_performance', 'athlete_average_speed']]
```

#Questions
1. Difference in speed for the 50k and 50mi between male and female

```python
final_df.groupby(['race_lenght', 'athlete_gender'])['athlete_average_speed'].mean()
```

2. what age groups are the best in the 50km race (20 + race min) show 20

```python
final_df.query('race_lenght == "50km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(20)
```

2. what age groups are the worst in the 50km race (20 + race min) show 20

```python
final_df.query('race_lenght == "50km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count>19').head(20)
```

3. when are runners slower in the summer or in the winter

```python
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2
```

```python
#change race_day column to datetime
final_df = final_df[final_df['race_day'].str.match(r'^\d{2}\.\d{2}\.\d{4}$', na=False)]
final_df['race_day'] = pd.to_datetime(final_df['race_day'], format='%d.%m.%Y')
```

```python
final_df['race_season'] = final_df['race_day'].dt.month.apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
final_df.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False)
```
